# Упражнения к главе 10
## Упражнение 10.1
Изменить пример в chap10.ipynb и убедиться, что дополнение нулями устраняет лишнюю ноту в начале фрагмента.

```python
from thinkdsp import read_wave
response = read_wave('code_180960__kleeb__gunshot.wav')
start = 0.12
response = response.segment(start=start)
response.shift(-start)
response.zero_pad(2**20)
response.normalize()
response.plot()
```

```python
violin = read_wave('code_92002__jcveliz__violin-origional.wav')
start = 0.11
violin = violin.segment(start=start)
violin.shift(-start)
violin.zero_pad(2**20)
violin.normalize()
violin.plot()
```

```python
spectrum = violin.make_spectrum()
transfer = response.make_spectrum()
output = (spectrum * transfer).make_wave()
output.normalize()
output.write("output.wav")
output.plot()
```

Действительно, лишняя нота в начале отсутствует.

## Упражнение 10.2
Смоделировать двумя способами звучание записи в том же пространстве, где была измерена импульсная характеристика, как сверткой самой записи с импульсной характеристикой, так и умножением ДПФ записи на вычисленный фильтр, соответствующий импульсной характеристике.

```python
slinky_response = read_wave('slinky_ir.wav')
slinky_response = slinky_response.segment(duration=5)
slinky_response.normalize()
slinky_response.plot()
```

```python
sound = read_wave("vla-bar-arp-bow-ed.wav")
sound = sound.segment(duration=5)
slinky_response.zero_pad(len(sound))
sound.shift(-0)
sound.normalize()
sound.plot()
```

```python
from thinkdsp import zero_pad
sound_spec = sound.make_spectrum()
slinky_transfer = slinky_response.make_spectrum()
output_mul = (sound_spec * slinky_transfer).make_wave()
output_mul.normalize()
output_mul.write("output_mul.wav")
output_mul.plot()
```

```python
output_conv = sound.convolve(slinky_response)
output_conv = output_conv.segment(0, 5)
output_conv.normalize()
output_conv.write("output_conv.wav")
output_conv.plot()
```

```python
Полученные результаты достаточно близки, однако умножение спектров добавляет лишнюю ноту в начале, с которой необходимо бороться путем добавления нулей.
```
