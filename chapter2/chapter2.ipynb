## Упражнение 2.2
Написать класс пилообразного сигнала, унаследованный от Signal

```python
from thinkdsp import Sinusoid
from thinkdsp import normalize, unbias
import numpy as np
class CustomSawtooth(Sinusoid):
    def evaluate(self, ts):
        full_cycles = self.freq * ts + self.offset / (2 * np.pi)
        remainder, frac = np.modf(full_cycles)
        return normalize(unbias(remainder), self.amp)

sawtooth = CustomSawtooth(freq = 440).make_wave(duration=0.05, framerate=40000)
sawtooth.plot()
```

Спектр сигнала:

```python
sawtooth_spectrum = CustomSawtooth(freq = 440).make_wave(duration=0.5, framerate=40000).make_spectrum()
sawtooth_spectrum.plot()
```

Сравнение со спектрами квадратного и треугольного сигналов (амплитуды уменьшены для лучшего сопоставления спектров):

```python
from thinkdsp import SquareSignal, TriangleSignal
square_spectrum = SquareSignal(freq = 440, amp = 0.5).make_wave(duration=0.5, framerate=40000).make_spectrum()
triangle_spectrum = TriangleSignal(freq = 440, amp = 0.78).make_wave(duration=0.5, framerate=40000).make_spectrum()
sawtooth_spectrum.plot(color='gray')
square_spectrum.plot()
```

```python
sawtooth_spectrum.plot(color='gray')
triangle_spectrum.plot()
```

Амплитуды гармоник пилообразного сигнала спадают аналогично гармоникам квадратного сигнала, однако их больше в 2 раза (присутствуют и четные, и нечетные гармоники). По сравнению с треугольным сигналом на пилообразном гораздо слабее затухают гармоники.

## Упражнение 2.5
Написать функцию, изменяющую все значения спектра обратно пропорционально частоте.

```python
from thinkdsp import Spectrum
def filter_spectrum(spectrum):
    new_spectrum = Spectrum(np.copy(spectrum.hs), np.copy(spectrum.fs), spectrum.framerate)
    new_spectrum.hs[0] = 0
    new_spectrum.hs[1:] /= new_spectrum.fs[1:]
    return new_spectrum

filter_spectrum(square_spectrum).plot()
```

```python
filter_spectrum(triangle_spectrum).plot()
```

```python
filter_spectrum(sawtooth_spectrum).plot()
```

Полученная функция по работе близка к ФНЧ, поскольку практически полностью отфильтровывает высокие гармоники и усиливает низкие частоты сигнала.

## Упражнение 2.6
Построить сигнал и с четными, и с нечетными гармониками, амплитуда которых уменьшается пропорционально квадрату частоты.

```python
from thinkdsp import SinSignal
freq = 440
freqs = np.arange(freq, 20000, freq)
amps = 1 / freqs ** 2
signal = sum([SinSignal(freq = fr, amp = a) for fr, a in zip(freqs, amps)])
signal.plot()
```

```python
signal.make_wave(framerate=40000).make_spectrum().plot()
```

При сложении косинусов вместо синусов форма сигнала отличается при аналогичном спектре:

```python
from thinkdsp import CosSignal
cos_signal = sum([CosSignal(freq = fr, amp = a) for fr, a in zip(freqs, amps)])
cos_signal.plot()
```

```python
signal.make_wave(framerate=40000).make_spectrum().plot(color='red')
cos_signal.make_wave(framerate=40000).make_spectrum().plot()
```
