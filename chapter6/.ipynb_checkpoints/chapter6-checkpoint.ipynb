## Упражнение 6.1
Убедиться в том, что analyze1 требует времени пропорционально n^3, а analyze2 - пропорциоально n^2 запуская их с несколькими разными массивами и засекая время работы.

```python
import time

def analyze1(ys, fs, ts):
    args = np.outer(ts, fs)
    M = np.cos(np.pi * 2 * args)
    amps = np.linalg.solve(M, ys)
    return amps

def analyze2(ys, fs, ts):
    args = np.outer(ts, fs)
    M = np.cos(np.pi * 2 * args)
    amps = M.dot(ys) / 2
    return amps

def benchmark_func(func, sizes):
    times = []
    for n in sizes:
        ts = (0.5 + np.arange(n)) / n
        fs = (0.5 + np.arange(n)) / 2
        ys = np.random.randn(n)
        start = time.time()
        func(ys, fs, ts)
        end = time.time()
        times.append(end - start)
    return times

sizes = np.array([100, 200, 500, 1000, 2000, 5000, 7500, 10000, 15000, 20000])
time_analyze1 = benchmark_func(analyze1, sizes)
time_analyze2 = benchmark_func(analyze2, sizes)
plt.figure()
plt.plot(sizes, time_analyze1, label='analyze1')
plt.plot(sizes, time_analyze2, label='analyze2')
plt.legend()
plt.show()
```

```python
plt.figure()
plt.loglog(sizes, time_analyze1, label='analyze1')
plt.loglog(sizes, time_analyze2, label='analyze2')
plt.legend()
plt.show()
print("analyze1", (np.log(time_analyze1[-1]) - np.log(time_analyze1[0])) / (np.log(sizes[-1]) - np.log(sizes[0])))
print("analyze2", (np.log(time_analyze2[-1]) - np.log(time_analyze2[0])) / (np.log(sizes[-1]) - np.log(sizes[0])))
```

Эксперимент показал, что в системе зависимости времени от размера в двойном логарифмическом масштабе практически параллельны (с точностью до случайных расхождений в начале). Это означает, что зависимости одного порядка, предположительно O(n^1.5)

## Упражнение 6.2
Реализовать версию ДКП в простей форме и применить его для записи музыки или речи. Определить сколько компонент можно удалить до того, как разница станет заметной.

```python
from thinkdsp import read_wave, decorate
gliss_wave = read_wave("153629_2626346-lq.wav")
gliss_dct = gliss_wave.make_dct()
gliss_dct.plot()
```

```python
def compress_dct(wave, threshold):
    dct = wave.make_dct()
    mask = np.abs(dct.amps) < threshold
    dct.hs[mask] = 0
    compressed_wave = dct.make_wave()
    return compressed_wave, dct

compressed_wave, compressed_dct = compress_dct(gliss_wave, threshold=30.0)
compressed_wave.normalize()
compressed_wave.write("compressed_wave.wav")
compressed_dct.plot()
```

На указанном пороге слышится существенно более отчетливый шум. Потери в %:

```python
remaining_amps = np.count_nonzero(compressed_dct.amps)
print((len(compressed_dct.amps) - remaining_amps) * 100.0 / len(compressed_dct.amps))
```
