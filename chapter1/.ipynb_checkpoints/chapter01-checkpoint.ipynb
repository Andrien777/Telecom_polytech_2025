## Упражнение 1.2
Скачать образец звука, имеющий четко выделенную высоту.
Был выбран сэмпл двух нот мелодики с сайта freesound

```python
from thinkdsp import read_wave
filename = "718970_15504031-lq.wav"
wave = read_wave(filename)
wave.plot()
```

Для анализа спектра необходимо извлечь фрагмент с приблизительно постоянной частотой, т.е. где проигрывается одна нота. Была выбрана вторая нота, по описанию ожидается C4.

```python
c4_start = 3.5
c4_duration = 3
segment = wave.segment(c4_start, c4_duration)
segment.plot()
```

Для вычисления спектра разработана функция make_spectrum, вычисляющая БПФ. Функция plot выводит график модулей значений БПФ.

```python
spectrum = segment.make_spectrum()
spectrum.plot()
```

Спектр практически полностью затухает к 8 КГц, для удобства анализа более высокие значения выводить смысла не имеет.

```python
spectrum.plot(high=8000)
```

На спектре наблюдаются пики, приблизительно равтоотстоящие друг от друга, что соответвует одной играемой ноте и ее гармоникам. Для поиска основной ноты необходимо найти частоту первого пика:

```python
list(filter(lambda x: x[1] < 1000, spectrum.peaks()))[0]
```

Данная частота приблизительно соответствует ноте C5, т.е. на октаву выше ожидаемой. Остальные пики наблюдаются на частотах, кратной данной.
При выделении ее из файла, высота тона не изменяется, но пропадет тембр, останется только звук частоты.

```python
lp_spec = spectrum
lp_spec.low_pass(600)
lp_spec.plot()
lp_wave = lp_spec.make_wave()
lp_wave.normalize()
lp_wave.apodize()
lp_wave.write("lp.wav")
```

Если эту частоту убрать, характер звука не изменится, но увеличится высота (примерно в 2-3 раза, до высоты следующего пика)

```python
spectrum = segment.make_spectrum()
hp_spec = spectrum
hp_spec.high_pass(600)
hp_spec.plot()
hp_wave = hp_spec.make_wave()
hp_wave.normalize()
hp_wave.apodize()
hp_wave.write("hp.wav")
```

Если отфильтровать самые громкие гармоники полосовым фильтром, в звуке остается некое подобие оригинального тембра, однако звучание изменяется практически до звучания чистой частоты.

```python
spectrum = segment.make_spectrum()
bs_spec = spectrum
bs_spec.band_stop(700, 3000)
bs_spec.plot()
bs_wave = bs_spec.make_wave()
bs_wave.normalize()
bs_wave.apodize()
bs_wave.write("bs.wav")
```

## Упражнение 1.3
Создание сложного сигнала из синуса и косинуса. Для синуса была выбрана частота 440 Гц с амплитудой 1, для косинуса - ее гармоника 880 Гц с амплитудой 0.5.

```python
from thinkdsp import SinSignal, CosSignal
sig_sin = SinSignal(440, 1)
sig_cos = CosSignal(880, 0.5)
sig = sig_sin + sig_cos
sig.plot()
```

Для дальнейшего анализа необходимо сигнал перевести в wave. При прослушивании сигнала ощущается только ведущая частота синуса.

```python
sig_wave = sig.make_wave(1)
sig_wave.normalize()
sig_wave.write("signal.wav")
```

При спектральном анализе обнаруживаются 2 пика на частотах, включенных в сигнал:

```python
sig_spectrum = sig_wave.make_spectrum()
sig_spectrum.plot(high = 1000)
```

```python
При добавлении частот, не кратных основной, они становятся различимы как отдельный тон. Были добавлены ноты E5 и E6 для звучания, соответствующему квинте Ля-Ми.
```

```python
sig_2 = sig + SinSignal(659.26, 1) + CosSignal(1319, 0.5)
sig_2.plot()
sig_wave_2 = sig_2.make_wave(1)
sig_wave_2.normalize()
sig_wave_2.write("signal2.wav")
```

Для анализа был вычислен спектр этого сигнала. Пики нот E появились на соответствующих частотах.

```python
sig_spectrum_2 = sig_wave_2.make_spectrum()
sig_spectrum_2.plot(high = 2000)
```

## Упражнение 1.4
Написать функцию stretch, растягивающую Wave на указанный коэффициент.
Для примера замедлим сигнал с нотами A4 и A5 в 2 раза. Получившийся сигнал должен содержать ноты A3 и A4.

```python
from thinkdsp import Wave
def stretch(wave: Wave, factor:float):
    return Wave(wave.ys, wave.ts * factor, wave.framerate / factor)

sig_wave_stretched = stretch(sig_wave, 2)
spectrum_stretched = sig_wave_stretched.make_spectrum()
spectrum_stretched.plot(high = 1000)
```

Пики сигнала приходятся на частоты 220 и 440 Гц, что и соответствует нотам A3 и A4. Значит функция работает верно.

```python
list(filter(lambda x: x[1] < 600, spectrum_stretched.peaks()))[0:2]
```
